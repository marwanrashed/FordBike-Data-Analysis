#### Add New columns for visualization purposes

```python
# Add a column for users ages, further age groups.
FordGoBike['age'] = 2019 - FordGoBike.member_birth_year
```

```python
# Add Age Groups column 
FordGoBike.loc[:,'Age_groups'] = pd.cut(FordGoBike['age'], bins = [18, 35, 55, 115], labels = ['Young Adults', 'Adults', 'Seniors'] )
```

```python
# Add columns for start time period of the day and for end_time too
FordGoBike['start_day'] = pd.DatetimeIndex(FordGoBike['start_time']).day
FordGoBike['end_day'] = pd.DatetimeIndex(FordGoBike['end_time']).day
```

```python
# Add columns for start time period of the day and for end_time too
FordGoBike['start_hour'] = pd.DatetimeIndex(FordGoBike['start_time']).hour
FordGoBike['end_hour'] = pd.DatetimeIndex(FordGoBike['end_time']).hour
```

```python
# Add start times of day column 
FordGoBike.loc[:,'Start_time_day'] = pd.cut(FordGoBike['start_hour'], bins = [-1,6,12,18,21,23], labels = ['Night', 'Morning', 'Afternoon', 'Evening','MidNight'] )
```

```python
# helper function
def MidNight_to_Night (x):
    if x == 'MidNight':
        return 'Night'
    else:
        return x
FordGoBike.Start_time_day = FordGoBike.Start_time_day.apply(lambda x: MidNight_to_Night(x))
```

```python
# Add end times of day column 
FordGoBike.loc[:,'End_time_day'] = pd.cut(FordGoBike['end_hour'], bins = [-1,6,12,18,21,23], labels = ['Night', 'Morning', 'Afternoon', 'Evening','MidNight'] )
FordGoBike.End_time_day = FordGoBike.End_time_day.apply(lambda x: MidNight_to_Night(x))
```

```python
# Ordering categorical types 
ordered_day_time = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ['Morning', 'Afternoon', 'Evening','Night'])
FordGoBike.Start_time_day = FordGoBike.Start_time_day.astype(ordered_day_time) 
FordGoBike.End_time_day = FordGoBike.End_time_day.astype(ordered_day_time) 
```

## Univariate Exploration

#### Firstly, add univariate plotting functions to avoid reptition and for more concise code.

```python
def line_plot (parameter, title, xlabel, ylabel, legend,fig_size = [8,5]):
    plt.figure(figsize = fig_size); 
    plt.plot(parameter, label = legend);
    plt.xlabel(xlabel);
    plt.ylabel(ylabel);
    plt.title(title);
    plt.legend();
    plt.show()
def ShowTwoVariable_lineplot (parameter1,parameter2, title, xlabel, ylabel, legend1,legend2,fig_size = [8,5]):
    plt.figure(figsize = fig_size); 
    plt.plot(parameter1, label = legend1);
    plt.plot(parameter2, label = legend2);
    plt.xlabel(xlabel);
    plt.ylabel(ylabel);
    plt.title(title);
    plt.legend();
    plt.show()
def Plot_Distribution (parameter,title, xlabel, ylabel, legend, fig_size = [8,5]):
    plt.figure(figsize = fig_size); 
    sns.distplot(parameter, label = legend);
    plt.xlabel (xlabel);
    plt.ylabel(ylabel);
    plt.title(title);
    plt.legend();
    plt.show()
def Plot_Hist (parameter,axis,title = None, xlabel = None, ylabel = None, legend = None,minimum = 0, maximum = 0, step = 0, fig_size = [8,5],r_width = 0.7,vis_x= True,vis_y = True,vis_legend = True):
    plt.figure(figsize = fig_size); 
    bins = np.arange(minimum, maximum, step)
    plt.hist(parameter, label = legend, bins = bins, rwidth = r_width);
    plt.title(title);
    plt.xlabel (xlabel).set_visible(vis_x);
    plt.ylabel (ylabel).set_visible(vis_y);
    plt.legend().set_visible(vis_legend);
    plt.axis(axis);
def Plot_HistLog (parameter,axis,title = None, xlabel = None, ylabel = None, legend = None,minimum = 0, maximum = 0, step = 0, fig_size = [8,5],r_width = 0.7,vis_x= True,vis_y = True,vis_legend = True):
    plt.figure(figsize = fig_size); 
    bins = 10 **np.arange(minimum, np.log10(maximum)+step , step)
    plt.hist(parameter, label = legend, bins = bins, rwidth = r_width);
    plt.title(title);
    plt.xlabel (xlabel).set_visible(vis_x);
    plt.ylabel (ylabel).set_visible(vis_y);
    plt.legend().set_visible(vis_legend);
    plt.xscale('log')
    plt.axis(axis);
def Plot_Count (parameter,title, xlabel, ylabel, legend, fig_size = [8,5]):
    plt.figure(figsize = fig_size);
    color_base = sns.color_palette()[0]
    sns.countplot(x = parameter, color = color_base, label = legend);
    plt.title(title);
    plt.xlabel (xlabel);
    plt.ylabel (ylabel);
    plt.legend();
    plt.show()
def Plot_bar (parameter1, parameter2, xlabel,ylabel,title,Rotation = 90):
    plt.figure(figsize = [15, 5]) 
    plt.bar(parameter1,parameter2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title);
    plt.xticks(rotation = Rotation)
    plt.show()
def Plot_barProp (parameter1, parameter2,Total_count, counts, xlabel,ylabel,title,fig_size = [8, 5],Rotation = 90):
    # rcParams['figure.figsize'] = 12,4
    plt.figure(figsize = fig_size) 
    plt.bar(parameter1,parameter2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title);
    locs, labels = plt.xticks(rotation=Rotation) 
    # loop through each pair of locations and labels
    for loc, label in zip(locs, labels):
        #count = counts[i]
        # Convert count into a percentage, and then into string
        pct_string = '{:0.1f}%'.format(100*count/Total_count)
        # Print the string value on the bar. 
        # Read more about the arguments of text() function [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.text.html)
        plt.text(loc, labels, pct_string, va='center', ha = 'center')
```

## Bivariate Exploration

Let's first specify the numeric/categorical features of interest in two separate lists.

```python
numeric_vars = ['duration_mins', 'age', 'start_day', 'start_hour']
categoric_vars = ['Start_time_day', 'Age_groups', 'user_type','member_gender','bike_share_for_all_trip']
```

In the dataset some features have both numeric and categroical representation such as: 
 - Age (numeric) - Age Group (Categorical)
 - Start_hour (numeric) - time of day (categorical)
 Only start day has a numeric representation and no categorical one. Let's first analyze the correlation between each of the numeric variables with the trip duration in mminutes using heat map.

```python
plt.figure(figsize = [8, 5])
sns.heatmap(FordGoBike[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

*The shown heatmap illustrates the linear correlation between each of the numeric variable and the others, including the variable of interest (duration in minutes). Unexpectedly, none of the numeric variables has a linear correlation with the trip duration. Nevertheless, they all have a very weak relationship.*

Despite the given heatmap, it will be also important to visualize the scatter plots of the numerical variables against the duration in mins. As in some occasions, there might not be a linear correlation, but a significant relationsfhip appears in the plots.
As shown in the univariate distribution, most of the durations were less than 50 mins. Hence, the bivariate distribution will be performed on the data lower than 60 mins for clarity.  

```python
samples = np.random.choice(FordGoBike.shape[0], 500, replace = False)
FordGoBike_samp = FordGoBike.loc[samples,:]
first_quartile = FordGoBike_samp.duration_mins >  FordGoBike_samp.duration_mins.quantile(0.25)
second_quartile = FordGoBike_samp.duration_mins < FordGoBike_samp.duration_mins.quantile(0.75)
one_hour_duration = FordGoBike_samp.duration_mins < 50
g = sns.PairGrid(data = FordGoBike_samp[one_hour_duration ], vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter);
```

*The figure above shows the scatter plot for all numeric values against the trips duration in minutes. It is clearly obvious that there is no any linear relationship between any of the features and the main variable of interest. However, we can say that with lower age the chances of having longer trip duration is higher. However, as in the univariate analysis, the days didn't have any obvious relationship with the trip duration, they all havesimilar chances. Similarly, the hours don't have clear relationship with the trip duration except in the night before sunrise, where there are no trips at all. I would like to investigate more the age feature vs. the trips duration using a heatmap.*

```python
bins_x = np.arange(-5, 85+1, 10)
plt.hist2d(FordGoBike_samp[one_hour_duration ].age ,FordGoBike_samp[one_hour_duration ].duration_mins  ,
           bins = bins_x)
plt.xlabel('Age');
plt.ylabel ('Trips Duration (minutes)')
plt.colorbar();
plt.show()
```

*The heat map of the age vs trips duration shows that most users are from young age. Additionally, having a higher trip duration also takes place mostly with younger users.*

The next step is to investigate the relationship between the categorical features and the trip duration. This multivariate analysis will take place using boxplots.  

```python
samples = np.random.choice(FordGoBike.shape[0], 2000, replace = False)
FordGoBike_samp = FordGoBike.loc[samples,:]
one_hour_duration = FordGoBike_samp.duration_mins < 50

def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sns.color_palette()[0]
    sns.boxplot(x, y, color = default_color)

plt.figure(figsize = [10, 10]);
g = sns.PairGrid(data = FordGoBike_samp[one_hour_duration ], y_vars = ['duration_mins'], x_vars = categoric_vars,
                size = 3, aspect = 1.5);
g.map(boxgrid);
# plt.ylim([-10, 40]);
plt.show();
```

From the boxplots we can see that some expected relationships weren't correctly anticipated, as some features don't have an affect at all. 
- Time of the day: It is shown that the morning and afternoon times have higher maximum trip duration. Yet, there medians and quartiles are so close to the night time at approximately 10 minutes. However, the evening time is slightly lower in the whole measures.
- Age groups: Similar to the time of the day feature, the three groups had medians lying around 10 minutes duration.
- User type: Despite the count of subscribers is much higher in the univariate analysis, the customers have more trip durations than them.
- Memeber gender: they are all so close to each other. Yet, the male group doesn't have a superiority here, as the female group leads slightly in the measures of spread.
- Bike share condition: they both are very similar to each other without any siginificant coclusion.

From the above box plots, I would like to take a closer look of the user type and memeber gender features using violin plot.

*The violin plot confirmed the concluded observations using the boxplot for the member gender feature.*
